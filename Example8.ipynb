```python
   from confluent_kafka import avro
   from confluent_kafka.avro import AvroConsumer
   from elasticsearch import Elasticsearch
```

```python
   # Kafka configuration
   kafka_config = {
       'bootstrap.servers': 'localhost:9092',
       'group.id': 'mygroup14',
       'schema.registry.url': 'http://localhost:8081',
       'auto.offset.reset': 'earliest'
   }
```

```python
   # Create AvroConsumer
   consumer = AvroConsumer(kafka_config)
   consumer.subscribe(['customer2topic'])
```

```python
   # Elasticsearch configuration
   es = Elasticsearch([{'host': 'localhost', 'port': 9200, 'scheme':'http'}])
```

```python
   # Consume messages and index to Elasticsearch
   while True:
        msg = consumer.poll(1.0)
        if msg is None:
           continue
        if msg.error():
           print(f"Consumer error: {msg.error()}")
           continue
        try:
            if msg.value():
                # Index message to Elasticsearch
                es.index(index='cust2python', body=msg.value())

                print(f"Indexed message: {msg.value()}")
            else:
                print("No data to index")  
        except ValueError as e:  
            print (f"Error indexing document: {e}")
```
